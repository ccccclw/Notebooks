### HW4

__Well-behaved functions requirements__:

- $\exists \int \psi^{*}(x) \psi(x) d x \space\space\space\space\space\space\space\space\space\space\space$   Normalizable
- $\exists ! \psi(x) \space\space\space\space\space\space\space\space\space\space\space\space\space\space\space\space\space\space\space\space\space\space\space\space\space\space\space\space$   Univalued
- $\lim _{x \rightarrow a} \psi(x)=\psi(a) \space\space\space\space\space\space\space$   Continuous
- $\lim _{x \rightarrow a} \frac{d \psi(x)}{d x}=\frac{d \psi(a)}{d x}\space\space\space\space\space\space$   Continuous first derivative

__Two operators commute if and only if they have the same eigenvectors.__

__Ladder Operator in Harmonic oscillator__: $
b=\left(\frac{m \omega}{2 \hbar}\right)^{\frac{1}{2}}\left(x+\frac{i p}{m \omega}\right), b^{\dagger}=\left(\frac{m \omega}{2 \hbar}\right)^{\frac{1}{2}}\left(x-\frac{i p}{m \omega}\right)
$
- __Eigenfunctions__ $b^{\dagger} \varphi_{n}(x)=\sqrt{n+1} \varphi_{n+1}(x)$

__Rotation in QM__: $\hat{L}_{x}=-i \hbar\left(y \frac{\partial}{\partial z}-z \frac{\partial}{\partial y}\right); \hat{L}_{y}=-i \hbar\left(z \frac{\partial}{\partial x}-x \frac{\partial}{\partial z}\right); \hat{L}_{z}=-i \hbar\left(x \frac{\partial}{\partial y}-y \frac{\partial}{\partial x}\right)$

__Rotation Ladder Operator__: $L_{+}=L_{x}+i L_{y};  L_{-}=L_{x}-i L_{y}$

2.$\space$ $\hat{H}=\hat{T}+\hat{V}(x) = -\frac{\hbar^{2}}{2 m} \frac{d^{2} }{d x^{2}}+\frac{k}{2} x^{2} ;\space \hat{V}(x)=\frac{k}{2} x^{2}$ 

$\hat{H}\hat{V}f = (-\frac{\hbar^{2}}{2 m} \frac{d^{2} }{d x^{2}}+\frac{k}{2} x^{2})\frac{k}{2} x^{2}f = -\frac{\hbar^{2}k}{4 m} \frac{d^{2} (x^{2}f)}{d x^{2}}+ \frac{k^2}{4} x^{4}f$

$\hat{V}\hat{H}f = \frac{k}{2} x^{2}(-\frac{\hbar^{2}}{2 m} \frac{d^{2} }{d x^{2}}+\frac{k}{2} x^{2})f = -\frac{\hbar^{2}k}{4 m} x^{2}\frac{d^{2} f}{d x^{2}}+ \frac{k^2}{4} x^{4}f$

Since $\frac{d^{2} (x^{2}f)}{d x^{2}} \neq  x^{2}\frac{d^{2} f}{d x^{2}}$, $[\hat{H}, \hat{V}] = \hat{H}\hat{V}- \hat{V}\hat{H} \neq 0$, so following the fact that "Two operators commute if and only if they have the same eigenvectors.", there is no eigenfunctions of the total energy that will also be eigenfunctions of the potential energy.

$\space$$\space$$\space$$\hat{L}_{y} \hat{L}_{z} f - \hat{L}_{z} \hat{L}_{y} f = -\hbar^{2}(z \frac{\partial f}{\partial y} -y \frac{\partial f}{\partial z} ) \rightarrow \left[\hat{L}_{y}, \hat{L}_{z}\right]=i \hbar \hat{L}_x$ 

By noticing the cyclic permutation of $x$, $y$, and $z$ in definition of $\hat{L}_{x}, \hat{L}_{y}$ and $\hat{L}_{z}$, we can replace $x$ by $y$, replace
$y$ by $z$, and replace $z$ by $x$, then we get $\space$$\space$$\space$$\hat{L}_{z} \hat{L}_{x} f - \hat{L}_{x} \hat{L}_{z} f = -\hbar^{2}(x \frac{\partial f}{\partial z} -z \frac{\partial f}{\partial x} ) \rightarrow \left[\hat{L}_{z}, \hat{L}_{x}\right]=i \hbar \hat{L}_y$
