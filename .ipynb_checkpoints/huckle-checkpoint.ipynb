```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.lines as mlines

a = input("Enter one number from 4,5,6 to get Huckel calculation \n result for Cyclobutadiene,Cyclopentadiene or Benzene: ")
n = int(a)
x = [np.around(-2*np.cos(2*np.pi*k/n), decimals=3) for k in range(1,n+1)]
print('('+'\u03B1'+'-'+'e'+')'+'/'+'\u03B2'+':',sorted(x))

#plot
if n == 4:
    fig, ax = plt.subplots(1,2,figsize=(4,2))
    circle = plt.Circle((0.5, 0.5), 0.4, color='b', fill=False)
    ax[0].add_artist(circle)
    points = [[0.5, 0.9], [0.1, 0.5], [0.5, 0.1], [0.9, 0.5]]
    line = plt.Polygon(points, fill=None, edgecolor='r')
    ax[0].plot(*zip(*points), marker='o', color='k', ls='')
    ax[0].add_artist(line)
    ax[0].axis('off')
    ax[1].axis('off')
    l = mlines.Line2D([0.4,0.6],[0.9,0.9])
    m = mlines.Line2D([0.2,0.4],[0.5,0.5])
    n = mlines.Line2D([0.6,0.8],[0.5,0.5])
    c = mlines.Line2D([0.4,0.6],[0.1,0.1])
    ax[1].add_artist(l)
    ax[1].add_artist(m)
    ax[1].add_artist(n)
    ax[1].add_artist(c)
    kwargs = {'length_includes_head' : True,'head_width':0.08, 'head_length':0.05,'shape':'right'}
    ax[1].arrow(0.46, 0, 0, 0.25,**kwargs)
    ax[1].arrow(0.54, 0.25, 0, -0.25,**kwargs)
    ax[1].arrow(0.26, 0.4, 0, 0.25,**kwargs)
    ax[1].arrow(0.66, 0.4, 0, 0.25,**kwargs)
    ax[1].text(0.85,0.475,r'$\alpha$')
    ax[1].text(0.8,0.07,r'$\alpha$'+'+'+r'2$\beta$')
    ax[1].text(0.8,0.87,r'$\alpha$'+'-'+r'2$\beta$')
    ax[0].text(0.18,-0.1,'Cyclobutadiene')
    ax[1].text(0.2,-0.18,'Energy level')
    plt.show()
    
elif n == 5:
    fig, ax = plt.subplots(1,2,figsize=(4.4,2))
    circle = plt.Circle((0.5, 0.5), 0.4, color='b', fill=False)
    ax[0].add_artist(circle)
    points = [[0.5-0.4*np.cos(7*2*np.pi/45), 0.5+0.4*np.sin(7*2*np.pi/45)], [0.5-0.4*np.sin(1*2*np.pi/5), 0.5-0.4*np.sin(1*2*np.pi/20)], [0.5, 0.1], [0.5+0.4*np.sin(1*2*np.pi/5), 0.5-0.4*np.sin(1*2*np.pi/20)],[0.5+0.4*np.cos(7*2*np.pi/45), 0.5+0.4*np.sin(7*2*np.pi/45)]]
    line = plt.Polygon(points, fill=None, edgecolor='r')

    ax[0].plot(*zip(*points), marker='o', color='k', ls='')
    ax[0].add_artist(line)
    ax[0].axis('off')
    ax[1].axis('off')
    l = mlines.Line2D([0.2,0.4],[0.85,0.85])
    j = mlines.Line2D([0.6,0.8],[0.85,0.85])
    m = mlines.Line2D([0.2,0.4],[0.37,0.37])
    n = mlines.Line2D([0.6,0.8],[0.37,0.37])
    c = mlines.Line2D([0.4,0.6],[0.1,0.1])
    ax[1].add_artist(l)
    ax[1].add_artist(j)
    ax[1].add_artist(m)
    ax[1].add_artist(n)
    ax[1].add_artist(c)
    kwargs = {'length_includes_head' : True,'head_width':0.08, 'head_length':0.05,'shape':'right'}
    ax[1].arrow(0.46, 0, 0, 0.25,**kwargs)
    ax[1].arrow(0.54, 0.25, 0, -0.25,**kwargs)
    ax[1].arrow(0.26, 0.27, 0, 0.25,**kwargs)
    ax[1].arrow(0.34, 0.52, 0, -0.25,**kwargs)
    ax[1].arrow(0.66, 0.27, 0, 0.25,**kwargs)
    ax[1].arrow(0.74, 0.52, 0, -0.25,**kwargs)
    ax[1].text(0.85,0.34,r'$\alpha$'+'+'+r'0.618$\beta$')
    ax[1].text(0.85,0.07,r'$\alpha$'+'+'+r'2$\beta$')
    ax[1].text(0.85,0.82,r'$\alpha$'+'-'+r'1.618$\beta$')
    ax[0].text(0.12,-0.1,'Cyclopentadiene')
    ax[0].set_xlim(0, 1)
    ax[0].set_ylim(0, 1)
    ax[1].text(0.25,-0.1,'Energy level')
    plt.show()
elif n == 6:
    fig, ax = plt.subplots(1,2,figsize=(4.4,2))
    circle = plt.Circle((0.5, 0.5), 0.4, color='b', fill=False)
    ax[0].add_artist(circle)
    points = [[0.5, 0.9], [0.5-0.4*np.sin(1*2*np.pi/6), \
            0.5+0.4*np.sin(1*2*np.pi/12)], [0.5-0.4*np.sin(1*2*np.pi/6), 0.5-0.4*np.sin(1*2*np.pi/12)], [0.5, 0.1],\
              [0.5+0.4*np.sin(1*2*np.pi/6), 0.5-0.4*np.sin(1*2*np.pi/12)], [0.5+0.4*np.sin(1*2*np.pi/6), 0.5+0.4*np.sin(1*2*np.pi/12)]]
    line = plt.Polygon(points, fill=None, edgecolor='r')

    ax[0].plot(*zip(*points), marker='o', color='k', ls='')
    ax[0].add_artist(line)
    ax[0].axis('off')
    ax[1].axis('off')
    l = mlines.Line2D([0.4,0.6],[0.9,0.9])
    m = mlines.Line2D([0.2,0.4],[0.3,0.3])
    n = mlines.Line2D([0.6,0.8],[0.3,0.3])
    p = mlines.Line2D([0.2,0.4],[0.6,0.6])
    q = mlines.Line2D([0.6,0.8],[0.6,0.6])
    c = mlines.Line2D([0.4,0.6],[0.1,0.1])
    ax[1].add_artist(l)
    ax[1].add_artist(m)
    ax[1].add_artist(n)
    ax[1].add_artist(p)
    ax[1].add_artist(q)
    ax[1].add_artist(c)
    kwargs = {'length_includes_head' : True,'head_width':0.08, 'head_length':0.05,'shape':'right'}
    ax[1].arrow(0.46, 0, 0, 0.25,**kwargs)
    ax[1].arrow(0.54, 0.25, 0, -0.25,**kwargs)
    ax[1].arrow(0.26, 0.2, 0, 0.25,**kwargs)
    ax[1].arrow(0.66, 0.2, 0, 0.25,**kwargs)
    ax[1].arrow(0.34, 0.45, 0, -0.25,**kwargs)
    ax[1].arrow(0.74, 0.45, 0, -0.25,**kwargs)
    ax[1].text(0.85,0.08,r'$\alpha$'+'+'+r'2$\beta$')
    ax[1].text(0.85,0.28,r'$\alpha$'+'+'+r'$\beta$')
    ax[1].text(0.85,0.58,r'$\alpha$'+'-'+r'$\beta$')
    ax[1].text(0.85,0.88,r'$\alpha$'+'-'+r'2$\beta$')
    ax[0].text(0.12,-0.1,'Cyclopentadiene')
    ax[1].text(0.25,-0.1,'Energy level')
    ax[0].set_xlim(0, 1)
    ax[0].set_ylim(0, 1)
    plt.show()
```
