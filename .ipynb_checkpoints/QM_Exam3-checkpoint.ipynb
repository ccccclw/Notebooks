$H_{i j}=\int f_{i} \hat{H} f_{j} d \tau$; $S_{i j}=S_{j i}=\int f_{i} f_{j} d \tau$

$H_{11}=\int_{0}^{l} f_{1} \hat{H} f_{1} d x = \frac{\hbar^2}{4m}(e^{-2l}-1)$; $S_{1 1}=\int_{0}^{l} f_{1} f_{1} d x = \frac{1}{2}(1 - e^{-2l})$

$H_{22}=\int_{0}^{l} f_{2} \hat{H} f_{2} d x = \frac{\hbar^2}{2m}(e^{-4l}-1)$; $S_{2 2}=\int_{0}^{l} f_{2} f_{2} d x = \frac{1}{4}(1 - e^{-4l})$

$H_{12}=\int_{0}^{l} f_{1} \hat{H} f_{2} d x = \frac{2\hbar^2}{3m}(e^{-3l}-1)$; $S_{1 2}=\int_{0}^{l} f_{1} f_{2} d x = \frac{1}{3}(1 - e^{-3l})$ 

$H_{21}=\int_{0}^{l} f_{2} \hat{H} f_{1} d x = \frac{\hbar^2}{6m}(e^{-3l}-1)$; $S_{1 2}=\int_{0}^{l} f_{1} f_{2} d x = \frac{1}{3}(1 - e^{-3l})$ 

$\left|\begin{array}{cc}H_{11}-E S_{11} & H_{12}-E S_{12} \\ H_{21}-E S_{21} & H_{22}-E S_{22}\end{array}\right|=0 \rightarrow $ 

$\frac{\hbar^{2}}{4 m}\left(e^{-2 l}-1\right) \frac{\hbar^{2}}{2 m}\left(e^{-4 l}-1\right)-(\frac{\hbar^{2}}{4 m}\left(e^{-2 l}-1\right) \left(-\frac{1}{4}\left(e^{-4 l}-1\right)\right)+\frac{\hbar^{2}}{2 m}\left(e^{-4 l}-1\right) \left(-\frac{1}{2}\left(e^{-2 l}-1\right)\right))E+\left(\frac{1}{4}\left(e^{-4 l}-1\right)\right)\left(\frac{1}{2}\left(e^{-2 l}-1\right)\right)E^{2}-\frac{\hbar^{2}}{6 m}\left(e^{-3 l}-1\right) \frac{2 \hbar^{2}}{3 m}\left(e^{-3 l}-1\right)+(\frac{\hbar^{2}}{6 m}\left(e^{-3 l}-1\right) \left(-\frac{1}{3}\left(e^{-3 l}-1\right)\right)+\frac{2 \hbar^{2}}{3 m}\left(e^{-3 l}-1\right) \left(-\frac{1}{3}\left(e^{-3 l}-1\right)\right))E-\left(-\frac{1}{3}\left(e^{-3 l}-1\right)\right)^{2}E^{2}=0$

$\left(\frac{\left(e^{-2 l}-1\right)\left(e^{-4 l}-1\right)}{8}-\frac{\left(\left(e^{-3 l}-1\right)\right)^{2}}{9}\right)\left(\frac{\hbar^{4}}{m^{2}}+\frac{5 \hbar^{2}}{2 m} E+E^{2}\right)=0$

$E = min\{-\frac{\hbar^{2}}{2m}, -\frac{2\hbar^{2}}{m}\} = -\frac{2\hbar^{2}}{m}$

As we know that $E > 0$, this variational result seems not to be a right bound. By just checking the condition of the wavefunction required for such system:

$P_0 = \lim _{x \rightarrow 0}\left(c_{1} e^{-x}+c_{2} e^{-2 x}\right)^{2}=\left(c_{1}+c_{2}\right)^{2} \neq 0$

$P_l = \lim _{x \rightarrow l}\left(c_{1} e^{-x}+c_{2} e^{-2 x}\right)^{2}=\left(c_{1} e^{-l}+c_{2} e^{-2 l}\right)^2 \neq 0$

we find that this is not an appropriate trial function.
