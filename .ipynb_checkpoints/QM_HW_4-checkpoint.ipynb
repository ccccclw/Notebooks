__Term__: A set of equal-energy atomic states that arise from the same electron configuration and that have the same $\mathbf{L}$ value and the same $\mathbf{S}$ value constitutes an atomic term.

__Total electronic angular momentum of an atom__: $\mathbf{J} = \mathbf{L} + \mathbf{S}$

__Level__: The set of states that belong to the same term and that have the same value of  $\mathbf{J}$
